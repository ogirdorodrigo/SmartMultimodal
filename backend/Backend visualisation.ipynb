```python
from scooter import find_nearest_tier
from nextbike import find_nearest_bikestations, find_nearest_freebikes
from weather import get_weather_description
```

```python
import folium
```

```python
start_coordinate = [52.453312, 13.384524]
m = folium.Map(location=start_coordinate, zoom_start=15)
folium.Marker(start_coordinate, popup='Here we are!').add_to(m)
m
```

```python
freebikes, dist_fbikes = find_nearest_freebikes(start_coordinate[1], start_coordinate[0])
[folium.Marker([coord[1], coord[0]], popup=f'I am a bike', icon=folium.Icon(color='green')).add_to(m) for coord in freebikes]
```

```python
station_bikes, dist_sbikes = find_nearest_bikestations(start_coordinate[1], start_coordinate[0])
[folium.Marker([coord[1], coord[0]], popup=f'{n} available bikes in this station', icon=folium.Icon(color='red')).add_to(m) for coord, n in station_bikes.items()]
```

```python
scooters, dist_scooters = find_nearest_tier(start_coordinate[1], start_coordinate[0])
[folium.Marker([coord[1], coord[0]], popup=f'I am a scooter', icon=folium.Icon(color='pink')).add_to(m) for coord in scooters]
```
